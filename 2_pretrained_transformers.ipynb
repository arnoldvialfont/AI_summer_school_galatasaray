### Document Classification

The first pipeline we can present is sentiment analysis, which is classified [among the '*text classification*' tasks](https://huggingface.co/tasks/text-classification):

```
classifier = pipeline("text-classification")
```

If we don't specify the model further, the library downloads the Distilbert model for sentiment analysis in English by default (`distilbert-base-uncased-finetuned-sst-2-english`). If relevant, we can then directly apply the classifier to an example and obtain the inferred label as well as a score that indicates the degree of confidence in the prediction:

```
classifier('Tranformers are super great !')
```

However, we can specify the model we want to download (provided it is suitable for the intended task). For example, in the ['models' section of the site](https://huggingface.co/models), by clicking on the `Text classification` and `French` tabs, we can find the `nlptown/bert-base-multilingual-uncased-sentiment` model, which is the most downloaded and is directly available in the `pipeline` object:

```
classifier = pipeline("text-classification", model = "nlptown/bert-base-multilingual-uncased-sentiment")

classifier("Transformers are not really great!")
```

### Token Classification

A second interesting pipeline in this library to present is that of named entity recognition (NER), which is integrated into the set of [token classification](https://huggingface.co/tasks/token-classification) tasks, including POS tagging.

```
classifier = pipeline("ner")
classifier("Hello I'm Arnold and I live in Paris.")
```

This time, the `dbmdz/bert-large-cased-finetuned-conll03-english` model is downloaded by default, but you can search for a model for a given language on the site. The most downloaded French model is `Jean-Baptiste/camembert-ner` (trained with Pytorch and available in the `pipeline` module) which reports very high F1 scores: PER = 0.9483, ORG = 0.8181, LOC = 0.8955, MISC = 0.8146. Several multilingual models are also available.

```
classifier = pipeline("ner", model = 'Jean-Baptiste/camembert-ner')
classifier("Bonjour, Je m'appelle Arnold et j'habite à Paris.")
```

To maintain a more readable output, it can be useful to specify the aggregation strategy so that sub-word tokens are grouped together (unless you want to train another algorithm afterwards):

```
classifier = pipeline("ner", model = 'Jean-Baptiste/camembert-ner', aggregation_strategy = "simple")
classifier("Bonjour, Je m'appelle Arnold et j'habite à Paris.")
```

### Question-answering (Tranformers versions prior to v.5.3.0 - March 2026)

Among document classifications, the Question Natural Language Inference (QNLI) task is concerned with determining whether a text contains the answer to a question. If we want to obtain the answer to a question, when it is a priori in the text, we apply a QA algorithm.

[In the transformers library](https://huggingface.co/tasks/question-answering), different variants of this task are available:
- extractive: when a context is available (text, table, html) for the algorithm which must detect the relevant passages;
- open and generative: a context is provided and the algorithm must generate a relevant answer on this basis;
- closed and generative: no context is provided and the algorithm must generate a relevant answer.

The classic evaluation score is then the F1 for the first variant or the exact match (based on the span of the returned characters).

```
qa_model = pipeline("question-answering")
question = "Where do I live?"
context = "My name is Arnold and I live in Paris."
```

```
qa_model(question = question, context = context)
```

For more information on the methods used to train such models, you can look at the [SQUAD 2 reference site](https://rajpurkar.github.io/SQuAD-explorer/) which provides datasets for training, presents the score of the same name for the different variants of this task and keeps a history of the best existing solutions.

For French, [Etalab offers a QA model](https://www.data.gouv.fr/fr/reuses/modele-de-questions-reponses-francophone/), `etalab-ia/camembert-base-squadFR-fquad-piaf`, available directly in the `pipeline` module:

```
qa_model = pipeline("question-answering", model='etalab-ia/camembert-base-squadFR-fquad-piaf')
question = "Where do I live?"
context = "Hello I am Arnold and I live in Paris."
```

```
qa_model(question = question, context = context)
```

Among the interesting variants of this QA task, we can also mention [Table Question Answering](https://huggingface.co/tasks/table-question-answering) for which the `google/tapas-base-finetuned-wtq` model is the most downloaded to date. There are none available in French at the moment.

### Translation (Tranformers versions prior to v.5.3.0 - March 2026)



Translation is one of the first tasks targeted by Transformer models, which are designed as encoder-decoders.

To date, the t5-base model is downloaded by default with the associated `pipeline`:

```
en_fr_translator = pipeline("translation_en_to_fr")
en_fr_translator("How old are you?")
```

However, it is common to instead use the models produced by the University of Helsinki, which offers more than 1400 models classified in language pairs, for example `Helsinki-NLP/opus-mt-en-fr`:

```
translator = pipeline("translation", model="Helsinki-NLP/opus-mt-en-fr")
translator("How old are you?")
```

In terms of score, the evaluation of the models is carried out with the BLEU and SACREBLEU scores [presented in this video from the Hugging Face course](https://youtu.be/M05L1DhFqcw).

### Text Summarization (Tranformers versions prior to v.5.3.0 - March 2026)

The model downloaded by default is `sshleifer/distilbart-cnn-12-6` for English. It is quite logically a seq2seq model, that is to say encoder-decoder.

```
classifier = pipeline("summarization")
classifier('''Paris is the capital and most populous city of France, with an estimated population of 2,175,601 residents as of 2018, in an area of more than 105 square kilometres (41 square miles). The City of Paris is the centre and seat of government of the region and province of Île-de-France, or Paris Region, which has an estimated population of 12,174,880, or about 18 percent of the population of France as of 2017.''')
```

The score used to evaluate the quality of the summary is the ROUGE score [presented in this video from the Hugging Face course](https://youtu.be/TMshhnrEXlg).

### Text Generation

For text generation, the model downloaded by default is `HuggingFaceTB/SmolLM3-3B`:

```
generator = pipeline('text-generation')
generator("Hello, I'm a language model", max_length = 30, num_return_sequences=3)
```

The text-generation pipeline can also be used for translation as follows:

```
translator = pipeline(
    "text-generation",
    model="Qwen/Qwen2.5-0.5B-Instruct"
)

```

```
messages = [
    {
        "role": "user",
        "content": (
            "Translate the following English text into French. "
            "My name is Arnold and I live in Paris."
        )
    }
]
```

```
translator(messages)
```

Navigating through the elements of the model response allows to extract the resulting part:

```
translator(messages)[0]["generated_text"][1]['content']
```

# Function calling for structured information extraction

A chat model may receive a list of available tools described by JSON schemas. In such a case, it can respond by generating a structured tool call with arguments to be parsed and executed in a subsequent code.

However, it may also be used to structured output generation:

```
pipe = pipeline(
    "text-generation",
    model="Qwen/Qwen2.5-1.5B-Instruct"
)

tool = {
    "type": "function",
    "function": {
        "name": "extract_decision_info",
        "parameters": {
            "type": "object",
            "properties": {
                "decision_number": {"type": "string"},
                "fine_amount_eur": {"type": "number"},
                "sanctioned_entities": {
                    "type": "array",
                    "items": {"type": "string"}
                },
                "leniency_applied": {"type": "boolean"}
            },
            "required": [
                "decision_number",
                "fine_amount_eur",
                "sanctioned_entities",
                "leniency_applied"
            ]
        }
    }
}

text = (
    "Decision 21-D-11 imposed a fine of 220 million euros on Company Macintosh. "
    "The authority did not apply leniency."
)

messages = [
    {"role": "user", "content": f"Extract the information: {text}"}
]

out = pipe(
    messages,
    tools=[tool],
)

out
```

```
out[0]['generated_text'][1]['content']
```

### Computer vision

The field of [computer vision](https://en.wikipedia.org/wiki/Computer_vision) is a branch of AI that relates to how a machine can acquire information from an image or video. The most powerful models are all based on complex neural networks, including [convolutional neural networks](https://en.wikipedia.org/wiki/Convolutional_neural_network), which we will not study here. Various pretrained models are available in the `transformers` library.

Image classification models are widely used for stock photography to assign a keyword to each image. The `image-classification` pipeline is initialized with the `google/vit-base-patch16-224` model by default. You then just need to specify a path, an http link or an image loaded in PIL. You can provide a `top_k` parameter which determines the number of results it should return (at the pipeline level or its call):

```
clf = pipeline("image-classification")
```

```
# https://en.wikipedia.org/wiki/Cat
clf(DATA_DIR + "Six_weeks_old_cat.jpg", top_k=1) # Kitten
```

By default, the model returns five:

```
clf(DATA_DIR + "Leopardus_guigna.jpeg") # Kodkod (Leopardus guigna)
```

### Audio

Automatic speech recognition (ASR), or Speech to Text (STT), consists of transcribing a given audio into text. It has many applications, such as voice user interfaces, and illustrates the type of tasks available in `transformers`.

The `pipeline` module notably contains a possibility of automatic speech transcription by specifying `automatic-speech-recognition`. The reverse operation is available by specifying `text-to-speech`.

```
pipe = pipeline("automatic-speech-recognition", model = "openai/whisper-large-v3")
```

```
# https://commons.wikimedia.org/wiki/File:Armstrong_Small_Step.ogg
pipe(DATA_DIR + "Armstrong_Small_Step.ogg")
```

### Inter-modality

Among the different multimodal solutions, we can mention Document Question Answering (or Document Visual Question Answering) which consists of answering questions about images of documents. DQA models take a (document, question) pair as input and return an answer in natural language. The models generally rely on features combining text, word position (bounding boxes) and the image.

This is useful when the question requires some understanding of the visual aspects of the document. However, some DQA models can work without images and allow you to ask questions based on spreadsheets, text PDFs, etc.!

Without specifications, of checkpoint, pipeline initializes `impira/layoutlm-document-qa`. We can also use `naver-clova-ix/donut-base-finetuned-docvqa` from Kim et al. (2021):

```
pipe = pipeline("document-question-answering", model="naver-clova-ix/donut-base-finetuned-docvqa")

question = "What is the purchase amount?"

pipe(image=DATA_DIR + "invoice-template-us-neat-750px.png", question=question)
```

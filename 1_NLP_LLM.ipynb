## From character strings to Vectors

To apply text mining and NLP algorithms, one must start with a **splitting of character strings into elements that can be analyzed** (having a minimal meaning here). This elements are called 'tokens, and the process is called "tokenization". Tokenization is often performed under the hood,but it can be a step to be explicitly configured/trained, depending on the language of the text or the specificity of the topics covered.

It is usual to perform this step sentence by sentence, so that **sentence segmentation**, **grouping sentences by paragraph** may be the first tasks, especially with long and particularly structured texts (e.g., legal texts).

Word tokenization can be done using the `SpaCy` library (the [reference website](https://spacy.io/) and the associated [Github repository](https://github.com/explosion/spaCy) provide good help on how to use it). In the vectorization process, the package `NLTK`  is useful ([reference website](http://www.nltk.org/) and associated [Github repository](https://github.com/nltk/nltk)) as well as `Scikit-Learn` ([reference website](https://scikit-learn.org/stable/) and associated [Github repository](https://github.com/scikit-learn/scikit-learn)).

### Normalization

This step is more or less demanding. Here, we will only present "stopwords" removal and lemmatization and noun chunk extractions.

- **Stopwords** correspond to irrelevant frequent words in a language in the sense that they do not help (by themselves) in understanding the content of the text. Removing them is immediate in `SpaCy` even though it might be better to proceed manually (depending on the analysis to perform).

````python
#SpaCy has a specific method associated to stopwords: .is_stop

for token in document[:10]:
    print("'{}' is a stopword: {}".format(token.text, token.is_stop))
````

````python
#Stopwords removal:
print([[word.text for word in sent if not word.is_stop] for sent in document.sents])
````

<div class="alert alert-info">
<b> Info:</b> It is possible to remove any punctuation equivalently with the <code>is_punct</code> method.    
</div>

````python
print(
    [[word.text for word in sent if not word.is_stop and not word.is_punct] for sent in document.sents]
)
````

- **Lemmatizing** corresponds to the process of using the 'lemma' of a word instead of the word itself. The main objective is to reduce the number of "tokens" in a corpus by regrouping them to unique words (both "am" and "are" are lemmatized to "be"). AGenAIn, this is immediate in `SpaCy` even though it might be better not to lemmatize texts (depending on the analysis to perform).

````python
#SpaCy has a specific method associated to lemmatization : .lemma_

for token in document[:10]:
    print("'{}' has for lemma: {}".format(token.text,token.lemma_))
````

````python
#Stopwords removal and lemmatizing:
print([[word.lemma_ for word in sent if not word.is_stop] for sent in document.sents])
````

<div class="alert alert-info">
<b> Info:</b> We see here that the <code>lemma_</code> method lowercases characters (except in tokens that are considered as proper nouns). In addition, in this example, this step does not seem crucial as most of lemmatized words are also stopwords. Other found lemmas are singular versions of initial words.
</div>

- **Chunking** noun propositions is an alternative (or eventually a complement) to the previous normalization tasks: instead of feeding an algorithm with vectors of words, it could be more appropriate to reduce the analysis to "noun chunks". This is performed with the `noun_chunks` method from `SpaCy`.

````python
#Stopwords removal and lemmatizing:
print(list(document.noun_chunks))
````

<div class="alert alert-info">
<b> Info:</b> We see here that this method asks for the use of the Python <code>list</code> method as it returns a <code>generator</code> and not directly a <code>list</code> object. Under the hood, this method performs "parsing" that is the grammatical analysis of words (using their <a href=https://en.wikipedia.org/wiki/Part-of-speech_tagging>part-of-speech tags</a>) as well as propositions (based on <a href=https://spacy.io/usage/linguistic-features#dependency-parse>specific word dependencies</a>). As an alternative, it is also possible to use the <code>NLTK</code> library and a context-free grammar in order to parse any chunks (see Chapter 7 section 2 and Chapter 8 in Klein et al. (2009), <a href=https://www.nltk.org/book>fully available here</a>).
</div>

### Vectorizing Texts

Simply speaking, this step consists in transforming lists of words in lists of numbers. It is the moment where we prepare the text for the algorithm.

Among many vectorization methods, we will limit ourselves here to the `One-hot` and `TF-IDF` encoding methods. They are the simplest ones to explain and mainly used for the basic steps of the paper. More methods are presented in Chapter 4 of Bengfort et al. (2018) ([available here](https://www.oreilly.com/library/view/applied-text-analysis/9781491963036/ch04.html)).

- **One-hot encoding** creates vectors of 0 and 1 for each document. The size of vectors is determined by the total vocabulary found in the corpus: a 0 is associated to the absence of a token in a document but that is found in the corpus ; a 1 indicates it is present in the document. The following figure illustrates this method assuming 3 documents in English (each made of only one sentence).

</figure>

`Scikit-Learn` is a simple solution to perform vectorization and convenient for following algorithm stpes. `CountVectorizer` is the dedicated object: it takes a list of strings as entry and yields a sparse matrix (see the [method presentation](https://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.CountVectorizer.html) and the [dedicated guide part](https://scikit-learn.org/stable/modules/feature_extraction.html#text-feature-extraction) for more information).

````python
#We replicate this example with a corpus in French made with the 3 sentences found in document
corpus=[sent.text for sent in document.sents]
corpus
````

````python
from sklearn.feature_extraction.text import CountVectorizer

#The binary option implies the use of 0 and 1 in the resulting vectors
vectorizer=CountVectorizer(binary=True) #add into braces for no lowercasing : , lowercase=False

#The .fit_transform() method applies the tasks associated to the Scikit-Learn object (here CountVectorizer)
vectors=vectorizer.fit_transform(corpus)

print("Corpus is made of {} documents and has a vocabulary of {} \n".format(vectors.shape[0],vectors.shape[1]))

#Once the vectorizer has been fitted and transformed, we can retrieve words in the vocabulary
print(vectorizer.get_feature_names_out(),'\n')

#A raw visualization can be obtained with the .toarray() method
print(vectors.toarray())
````

<div class="alert alert-info">
<b> Info:</b>  We see here that  <code>Scikit-Learn</code> implicitely tokenizes the text (with the <code>build_analyzer()</code> method  <a href=https://scikit-learn.org/stable/modules/feature_extraction.html#common-vectorizer-usage> that extracts words of at least 2 letters</a>), and lowercases characters. Lowercasing can be disabled by adding <code>lowercase=False</code> next to <code>binary=True</code>. Removing <code>binary=True</code> performs an other encoding, namely <strong>term-frequency</strong> (as compared to one-hot encoding: 1s are replaced by the number of times a token appears in the document).
</div>

It is possible to use the tokenizer of `SpaCy` instead of the one implemented by the `CountVectorizer`, as well as all methods trained in several languages (stopwords and punctuation removals, lemmatizing...). This first asks for a simple workaround consisting in specifying a function (named `identity`) that returns the submitted object. The object in question here is the sentence tokenized by `Spacy`.

````python
#Some function that returns the entered object
def identity(sentence):
    return sentence
````

````python
#We create a list of the documents tokenized by SpaCy
corpus_spacy=[[word.lemma_ for word in sent if not word.is_stop and not word.is_punct] for sent in document.sents]
print(corpus_spacy)
````

````python
#This second vectorizer specifies no lowercasing (if wanted, must be done above) and the tokenizer is disabled de facto

vectorizer2=CountVectorizer(binary=True, lowercase=False, tokenizer=identity)

vectors2=vectorizer2.fit_transform(corpus_spacy)


print("Corpus is made of {} sentence(s) and has a vocabulary of {} \n".format(vectors2.shape[0],vectors2.shape[1]))

#Once the vectorizer has been fitted and transformed, we can retrieve vocabulary
print(vectorizer2.get_feature_names_out(),'\n')

#A raw visualization can be obtained with the toarray() method
print(vectors2.toarray())
````

- **TF-IDF encoding** only departs from the one-hot encoder in that it uses real numbers instead of 1s in each document. This real value is calculated for a token as the product of **term frequency** (TF) and **inverted document frequency** (IDF). The following figure illustrates this method with 3 documents in English.
<figure>
    <div align="center" style="margin-top:0.5cm"> <b> Vectorization with "TF-IDF" encoding and lemmatizing (Bengfort et al., 2018, p.62)</b> <br>
  <img src="https://github.com/arnoldvialfont/AI_summer_school_galatasaray/blob/master/img/vect_tfidf.png?raw=1" style="margin-top:0.5cm;margin-bottom:0cm;width:80%"></div>
</figure>

In practice, IDF is calculated in a given document by the formula:
$$
IDF(\text{token}) = log\left(\frac{n}{DF(\text{token})}\right),
$$

where $n$ is the number of documents in the corpus and $DF(\text{token})$ is the number of documents in which the token appears at least once.

The `Scikit-Learn` implementation of the TF-IDF method is straightforward, even though some practical modifications are introduced in the above formula as well as some normalization ([see the documentation for more information](https://scikit-learn.org/stable/modules/feature_extraction.html#tfidf-term-weighting)).

````python
from sklearn.feature_extraction.text import TfidfVectorizer

#We can integrate the result from SpaCy tokenizer as follows

vectorizer3=TfidfVectorizer(binary=True, lowercase=False, tokenizer=identity)

vectors3=vectorizer3.fit_transform(corpus_spacy)


print("Corpus is made of {} sentence(s) and has a vocabulary of {} \n".format(vectors3.shape[0],vectors3.shape[1]))

#Once the vectorizer has been fitted and transformed, we can retrieve vocabulary
print(vectorizer3.get_feature_names_out(),'\n')

#A raw visualization can be obtained with the toarray() method
print(vectors3.toarray())
````

## Embedding and Context

Pre-trained models typically use vectorizations based on semantic spaces, where the distribution of words and therefore documents are distributed by vectors whose coordinates describe their meaning.


One of the steps that has proven essential in the emergence of LLMs and Generative AI (GenAI herafter) has been the use of **self-supervised learning** in which the text itself is used to create learning tasks. This method has been both a way to circumvent the difficulties and the cost of human annotation of large quantities of textual data, and an opportunity to take advantage of the wealth of data available on the Internet and social networks. This is also one of the reasons that explains the omnipresence of GAFAM companies (Google, Amazon, Facebook, Apple, Microsoft) and other economic actors in the digital sector in the development of these models: they have long had massive quantities of textual data from Internet searches, customer reviews, social networks, etc.


One of the first models trained in this direction is word2vec, created by researchers at Google (Mikolov et al., 2013). In Continuous Bag of Words (CBOW): the aim is to predict the word at the center of a given sequence or "context" (using a neural network). Here we have the idea of constructing a probabilistic language model from the left and right towards a word in the center of the context.
<div>
<div align="center" style="margin-top:0.5cm"> <b> Unlabeled texts becomes alternately an example or a label. (Vajjala et al., 2020, p.98)</b><br>
<img src="https://github.com/arnoldvialfont/AI_summer_school_galatasaray/blob/master/img/embedding_cbow.png?raw=1" style="margin-bottom:0.5cm;width:70%"></div>
</div>


Self-supervised training, as its name suggests, is characterized by the fact that the data is the very support for the creation of learning tasks. In terms of NLP, we find for example at the center of this approach the *masked word prediction*. For example, in the sentence "The cat is sleeping on the ___", the model must learn to predict that the missing word is probably "table" or "chair" rather than "car" or "forest". By learning to perform this task, the model develops a logical representation of natural language and its underlying structures to classify words relative to each other. More precisely, these methods result in words that are themselves represented by vectors, whose coordinates reflect their degree of co-occurrence in similar sentences, and capture the semantic relationships of similarity or distance.[See Smith (2020) for a complete and accessible presentation of this approach, from binary encoding representations to matrices representing each document.]

It is still possible to present here *word2vec*, one of the first models of this type presented in Mikolov et al. (2013) based on a shallow neural network architecture (*shallow learning*). This model popularized the idea that words could be represented by vectors in a reduced dimension space $N$, between 50 as presented in Pennington et al. (2014) to several thousands in some models. The following figure, from Rong (2016), illustrates the principle of this training according to the so-called *continuous bag of words* (CBOW), on the left, and *skip-gram*, on the right, approaches. The two approaches respectively concern the prediction of words *at the center* or *around* a token in a vocabulary of size $V$, and using a context of a number of words $C$. Concretely, the words are encoded by vectors of size $V$ in a binary way before training and passed through a neural network whose parameters are adjusted to minimize prediction errors at the output.

![word2vec training models: CBOW vs Skip-Gram](https://github.com/arnoldvialfont/AI_summer_school_galatasaray/blob/master/img/CBOW_Skip-gram.png?raw=1)

After self-supervised training, the matrices $W$ and $W'$ are recovered because they contain the vector representations of the words in the vocabulary. More precisely, these are the matrices of the parameters of the neural network, whose respective sizes $V \times N$ and  $N \times V$ are defined by the size of the vocabulary $V$ and the number of neurons $N$ which is chosen before training. It is then possible to associate coordinates with each word to perform a *word embedding* in a **semantic space** that is no longer based solely on the vocabulary but on the way words are interchangeable.

In addition, it is then possible to **construct logical operations with words**. Indeed, the different dimensions can be seen as representing systematic variations of language, for example gender, verb tense or the political relationship of cities and countries. We can for example verify the results of the following vector calculations: $king - man + woman \simeq queen,\text{ } walked - walking + swimming \simeq swam, \text{ or } Spain - Madrid + Berlin \simeq Germany$. This is illustrated in the following figure where it is effectively presented paths between two words relatively parallel in a given dimension.

![Semantic dimensions in word2vec (image from developers.google.com)](https://github.com/arnoldvialfont/AI_summer_school_galatasaray/blob/master/img/word_embedding.png?raw=1)

It is important to note that word embedding models like *word2vec* have since been surpassed by more complex architectures based on deep neural networks (*deep learning*) that we do not present. However, the fundamental principle of self-supervised training remains relevant for what follows: learning from large amounts of unannotated textual data for a logical understanding of the observed natural language.

<div class="alert alert-danger">
<b> Warning:</b> Two remarks are necessary here, although they are valid for several machine learning contexts:

- The training of embeddings must be done on a very large corpus to obtain satisfactory results (i.e., word vectors). Whenever possible, pre-trained algorithms are used rather than training "from scratch" because their construction is very costly in terms of time and expense, especially energy. - If the loaded model does not end with `sm` (i.e., is not the *small* version of the language), we can access its vocabulary using the `.vocab` method. In this case, we have access to the [`.most_similar`](https://spacy.io/api/vectors#most_similar) method to find the words closest to another:

````python
word = 'king'

ms = 10

print('For the word :', word)

nlp_lg = spacy.load('en_core_web_lg')

vec_spacy = nlp_lg.vocab[word].vector

result = nlp_lg.vocab.vectors.most_similar(vec_spacy.reshape(1,-1), # il faut une matrice : np.array([1,2,3]).reshape(1,-1) = np.array([[1,2,3]])
                                              n=ms)

words = [nlp_lg.vocab[i].text for i in result[0][0]]

print(f'\nClosest {ms} words are : \n\n', words[1:]) #La méthode renvoie en premier le mot donc on enlève words[0]
````

- We can consider simple operations:

````python
#Download step (needed once in a local environment)
!python -m spacy download fr_core_news_lg

nlp_fr_lg = spacy.load('fr_core_news_lg')

words_list = [('roi', 'homme', 'femme'),

              ('marchant', 'marcha', 'nagea'),

              ('Espagne', 'Madrid', 'Berlin'),
]

for word_triple in words_list :

    vec_spacy1 = nlp_fr_lg.vocab[word_triple[0]].vector

    vec_spacy2 = nlp_fr_lg.vocab[word_triple[1]].vector

    vec_spacy3 = nlp_fr_lg.vocab[word_triple[2]].vector

    vec_spacy4 = vec_spacy1 - vec_spacy2 + vec_spacy3

    result4 = nlp_fr_lg.vocab.vectors.most_similar(vec_spacy4.reshape(1,-1), n=ms)

    words4 = [nlp_fr_lg.vocab[j].text for j in result4[0][0]]

    print(f"\n- Operation '{word_triple[0]}' - '{word_triple[1]}' + '{word_triple[2]}' gives : '{words4[1]}'.")
````

- By studying the proximities of the vectors associated with the documents, we can quite simply measure a form of semantic similarity degree (with the `cosine` method from `scipy`):

````python
from scipy.spatial.distance import cosine
````

````python
# Dowload the small model for French:
!python -m spacy download fr_core_news_sm
nlp_fr_sm = spacy.load("fr_core_news_sm") #17.2MB
documents = ["j'aime la peinture", "je fais du basketball", "j'aime le basketball"]
v1 = nlp_fr_sm(documents[0]).vector.reshape(-1)
v2 = nlp_fr_sm(documents[1]).vector.reshape(-1)
v3 = nlp_fr_sm(documents[2]).vector.reshape(-1)

print('distance v1-v2:',cosine(v1,v2))
print('distance v1-v3:',cosine(v1,v3))
print('distance v2-v3:',cosine(v2,v3))
````

## Large Language Models (LLM)

Once the notion of the relative placement of words in a semantic space has been integrated into the modeling of natural language, it is possible to address the importance of the **context in which words appear**. Words indeed have a certain fluidity of meaning, depending on the sentences, the syntactic structure or even the sender of the message. For example, the word "livre" in French can refer to a physical object (a book), an action (associated with the verb to deliver) or a monetary unit (English, Turkish or else). In human language, it is then contextual clues that potentially allow their exact meaning to be determined: the appearance of another word like "bibliothèque" (library) which points to the meaning of a book or a possible specific sequence with a pronoun like "je" (I) before the word (indicating the action of a delivery). However, the use of common sense can be implicit and it is not certain that there will be a specific modulation of the representation of a word in the trained model.

This is where the construction of LLMs begins, which aim to capture these contextual nuances by using deep neural network architectures, notably *transformers* (Vaswani et al., 2017). These models were not the first architectures used to capture context, but they have allowed the most significant advances in terms of performance and computational efficiency. Their main innovation lies in the *self-attention mechanism*, which allows the model to learn the modulation of the coordinates of words in the semantic space according to those of the other words present in a sentence. Thus, during self-supervised training on a corpus, each word is linked to the others and the learning focuses on their degree of contextual dependence. The following figure, from Tunstall et al. (2020), illustrates this principle by showing with bold lines that the vector of the English word "flies" is particularly modulated by the vectors of the words "arrow" and "banana": the word on the left will have coordinates close to those of the word "vole" (flies, as in "time flies") and the one on the right to those of the word "mouches" (flies, as in the insect).

![](https://github.com/arnoldvialfont/AI_summer_school_galatasaray/blob/master/img/attention-weights-fruits.png?raw=1)

The model illustrated in this figure is BERT (Devlin et al., 2019), which was one of the first LLMs to demonstrate substantial performance improvements in a variety of natural language processing tasks.
````python
from scipy.spatial.distance import cosine
````

````python
documents = ["I like painting", "I play basketball", "I like basketball"]

v1=model(**tokenizer(documents[0], return_tensors="pt"))[0].mean(axis=1).detach().numpy().reshape(-1)
v2=model(**tokenizer(documents[1], return_tensors="pt"))[0].mean(axis=1).detach().numpy().reshape(-1)
v3=model(**tokenizer(documents[2], return_tensors="pt"))[0].mean(axis=1).detach().numpy().reshape(-1)

print('distance v1-v2:',cosine(v1,v2))
print('distance v1-v3:',cosine(v1,v3))
print('distance v2-v3:',cosine(v2,v3))
````

## Models for Instructions

Generative Artificial Intellegence (GenAI) has its origins in the category of LLMs trained to generate text (or other objects) in response to **instructions** given by a user. For exemple, Ouyang et al. (2022) presents InstructGPT:
- a model with a GPT-like architecture is subjected to supervised fine-tuning to learn coherent text generation with a dataset of pairs ($instruction$, $expected\, response$),
- then is refined by a reinforcement learning step from human feedback on the generated texts (*Reinforcement Learning from Human Feedback*, RLHF).

In this last step, human evaluators rate the model's intermediate responses according to their preferences, and these ratings are used to train a reward model that guides the optimization of the main model.[See Zhang et al. (2025) for a literature review on LLM alignment methods, and in particular reinforcement methods where it is the model's experiments and the associated rewards that make it optimize its parameters.]

Although other methods are developing rapidly as an alternative to RHLF, potentially more efficient, it is important to note here a third layer of training compared to the first LLMs: the alignment of the model's responses to a user's expectations. The context of text generation is then potentially derived from conversational interactions in **prompts** which are seen as textual instructions. These instructions can be more or less long and complex, include examples, constraints of form or mode of response. However, the model's ability to understand and follow these instructions depends on the quality of the initial training and the infrastructures that are made available to the **conversational agent**: computing power and time, access to external information resources, implicit use of other agents, etc. Note here that the prompts are not used for training the model in the context of the current discussion, but rather to guide the generation of text in real time according to the overall objective of the model.

The following example illustrates a truncated prompt to understand the difference between a standard LLM and a model adjusted for GenAI: the latter expects a more complete instruction to generate an adapted response, possibly by asking for clarifications to achieve the objective.

| **Prompt** | **Response from a GPT-type LLM** | **Response from an instruction-oriented LLM** |
|------------|-------------------------------|---------------------------|
| "Paris is the" | "Capital of France." | "It seems you forgot to complete your sentence, are you looking for the capital of France?" |


A major challenge for GenAI then appears: the ability of models to align with instructions formulated in natural language. Two aspects are in fact to be considered, namely the resources necessary for the formulation of the responses and the effective degree of alignment with the user's needs. These two points are presented in the rest of this contribution in two stages, first considering commercial online solutions and then open source solutions that are potentially local, highlighting their respective advantages.

## Online vs. Local Services

This section presents the main solutions for mobilizing GenAI, starting with online platforms, then presenting local solutions. Faced with the variety of available solutions, it was chosen to present solutions representative of the dominant approaches by highlighting their advantages and limitations for teaching and university research. For a more exhaustive comparison, oriented towards economic applications and regularly updated, see Korinek (2023, 2024 and 2025) and the companion site of this series of articles entitled Generative AI for Economic Research (https://www.genaiforecon.org/).

### Commercial solutions

First of all, it should be specified that commercial solutions for GenAI are generally based on online platforms that offer user-friendly interfaces and advanced language models. In addition, these platforms offer additional features, often for a fee, such as response customization, integration with other tools and services, and various automation solutions.

However, to simply present the performance of these solutions, we must go back to the basis of these services: the language models that underlie them. The majority of platforms offer different versions of their models, one of which, more complex and powerful, serves as a reference in institutionalized and systematically updated competitions. The LMArena site (https://lmarena.ai/) presented in Chiang et al. (2024) is an illustration where the models are compared in pairs by users and ranked according to the Elo score (from chess player competitions). The following figure (proposed by McCandless, Evans and Barton on the [informationisbeautiful.net](https://informationisbeautiful.net/visualizations/the-rise-of-generative-ai-large-language-models-llms-like-chatgpt/) site) presents the scores of the leaders of LMArena for textual instructions in September 2025 and the SWE-Bench score which measures their performance on solving real programming problems. Open source models are indicated by a diamond.

![SWE-Bench and Elo scores (figure)](https://github.com/arnoldvialfont/AI_summer_school_galatasaray/blob/master/img/swe-bench_lmarena.png?raw=1)

This graph shows the link between subjective user preferences (upwards) and more objective scores of the quality of the GenAI models' responses (to the right). It first indicates a strong positive correlation between these two scores, meaning that users are able to judge the quality of the generated responses quite reliably. In addition, we can see that the dominant commercial solutions are also those that obtain the best scores in objective evaluations: these platforms are able to provide high-quality responses to users.[Different metrics exist to evaluate LLMs. Another, more general one, called Massive Multitasks Language Understanding (MMLU), ranks the models differently but is subject to criticism, particularly since an April 2024 article in the New York Times, published by Kevin Roose, according to which the training would be carried out specifically to optimize this score. See the page https://lifearchitect.ai/datasets-table/ to access the table of scores illustrated in the graph and the presentation of the different metrics used (consulted in October 2025).]

These high performances are made possible by access to massive computing infrastructures, allowing for the training and deployment of very complex language models. It should be noted that these two stages require considerable computing power and significant energy resources: excessive use is possible compared to what would be *collectively optimal*. In addition, these platforms benefit from training data that contribute to their superiority in terms of performance: free access for users is compensated in the general conditions of use by the right to collect interaction data, which are then used to refine and improve the language models during their updates.

### The limits and inherent risks of online platforms

Different problems arise in relation to the question of model alignment:

- they are used independently of the user's local environment,

- the data provided in the context of the requests are exposed to a commercial company, and

- a risk of lock-in to a single solution appears.

The first major problem appears in the fact that the models are online and **do not always adapt to the context in which the user formulates a request**. Indeed, the responses are, at least at the beginning of a discussion thread with the conversational agent, relatively standardized and may require several exchanges or new attempts before reaching the expected result. This inefficiency in use comes from the fact that the documents or data necessary for a relevant response cannot always be provided to the model, due to their unavailability, number or size, but also that the model does not know a priori the origin of the request.


A possible solution to this problem is the use of locally embedded versions of these tools. Such a solution is made possible with, for example, the installation of software such as Visual Studio Code (free) and a subscription to the GitHub platform (free for students and academic staff). In such a framework, it is then sufficient to install a few extensions to the default version of Visual Studio Code to be offered to use a wide variety of models from different platforms (GPT(s), Claude, Gemini, etc.). To date, three types of interactions are planned: answering questions, editing documents in the workspace, but also the agent mode which gives the model a wide freedom of action on the computer.

However, a second more fundamental problem comes from the fact that any use of GenAI implies the the conversational agent has **access to individual or professional data provided by the user**. This second point is even reinforced by the solution mentioned in response to the previous one, which can make it impossible to choose in a research context for example. Here, it is a reinforced contractualization with the service provider that can be chosen (including a confidentiality clause) or, conversely, the use of a purely local solution (without calling on commercial solutions). This point is the subject of the next section.

Finally, in the event that the use of a proprietary GenAI proves compatible with a given use, a final risk is associated with the form of **habituation or even lock-in** that may result. Indeed, if it seems that some models are the most efficient at a certain date, it is important to point out the instability of the rankings and the specific investment that can be made in a given solution.[ However, OpenAI, for example, was among the first providers to develop the form constraints of the generated outputs, with the JSON and Function calling modes. The redeployment of a project for another platform can then prove to be complicated. This risk is lower with the hybrid solution presented above.] A possible lock-in effect is a source of inefficiency known to economists on the markets, particularly through the reduction of competition it generates: the first entrant benefits from the cost of *switching* between services See Korinek and Vipra (2024) for a discussion of this point in the context of GenAI.



It should also be noted that this risk is compatible with the known use by some economic actors of the *hypernudge*. This practice consists in guiding users' choices with the help of AI by exploiting their cognitive biases: possible orientation in the comparison of solutions to a problem (See Richarde et al., 2025), maintenance in the discussion after reaching the instruction, confirmation bias (Fanous et al., 2025), political bias (Motoki et al., 2025). These risks in GenAI interfaces are possible as soon as the objective of the service provider is not necessarily aligned with that of the user.

### Non-commercial and local solutions

In the context of teaching and university research, it is important to keep in mind this question of independence from service providers, particularly for ethical reasons and data confidentiality. The report of the Ministry of Higher Education, Research and Space (2025) on AI and higher education in France insists on this point by recommending, for example, the use of open source and local solutions when possible. As an alternative to local solutions, the use of secure APIs can be considered, including OpenGateLLM opened to agents of the State public service in January 2025 by Etalab, which is the AI department of DINUM (https://github.com/etalab-ia/OpenGateLLM).[The AI Technical Day of September 25, 2025 organized by Esup, whose video is available online, allows to see concrete examples of feedback from AI solutions at the University: https://chk.me/t74F5am.]

To date, several universities and research laboratories have developed local solutions to mobilize GenAI for the purpose of assisting users, for example in documentary research (RAG for Retrieval-Augmented Generation). These often rely on the use of open source models, such as LLaMA (Meta), Mistral (Mistral AI) or Qwen (Alibaba), and a comparison of their performance can be made in conjunction with the competition hosted by Hugging Face (Open LLM Leaderboard).

These solutions can be deployed on internal IT infrastructures or on secure cloud computing services, but lighter solutions can also be considered **provided you have access to a GPU**. For a minimalist need to simply ask questions to open source LLMs, many software are available such as LMStudio (https://lmstudio.ai/) or Openweb UI (https://openwebui.com/). These two solutions allow you to load open source language models and query them via a user-friendly graphical interface. An important limitation to point out, however, remains the risk of hallucinations, that is, the generation of incorrect or unfounded answers, which can be more pronounced in these models compared to commercial solutions benefiting from additional training and regular updates.

In addition, one of the major advantages of these local solutions is the **confidentiality of the data** and the **increased personalization** of the models. Indeed, sensitive or confidential data do not leave the user's local environment, which reduces the risk of leakage or misuse by third parties. They can therefore be adapted to the specific needs of students or the research field. However, this requires investments in time and resources for the installation, configuration and maintenance of the models and associated infrastructures that it is impossible to describe here.
